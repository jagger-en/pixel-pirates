```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_normalized_data(csv_file, X, Y, x1, y1, N):
    # Read data from CSV file into a DataFrame
    data = pd.read_csv(csv_file)

    # Extract desired columns for plotting
    timestamp = data['Timestamp']
    x_velocity = data[X]
    y_velocity = data[Y]
    x_destination = data[x1]
    y_destination = data[y1]

    # Calculate the magnitude (normalized velocity) for each data point
    normalized_velocity = np.sqrt(x_velocity**2 + y_velocity**2)

    # Calculate the magnitude (normalized destination) for each data point
    normalized_destination = np.sqrt(x_destination**2 + y_destination**2)

    # Create the plot
    plt.figure(figsize=(10, 6))
    plt.plot(timestamp, normalized_velocity, marker='o', linestyle='-', label='Normalized Velocity')
    plt.plot(timestamp, normalized_destination, marker='x', linestyle='-', label='Normalized Destination')

    # Add labels, legend, and title
    plt.xlabel('Timestamp')
    plt.ylabel('Normalized Value')
    plt.title(f'Normalized Velocity and Destination vs. Timestamp {N}')
    plt.legend()

    # Show the plot
    plt.grid(True)
    plt.show()
```

```python
 csv_file = 'Dev_data.csv'
```

```python
#Obj1
plot_normalized_data(csv_file, '1stOSp_X [m/s]', '1stOSp_Y [m/s]','1stODist_X [m]','1stODist_Y [m]', 1)
```

```python
#Obj2
plot_normalized_data(csv_file, '2ndOSp_X [m/s]', '2ndOSp_Y [m/s]','2ndODist_X [m]','2ndODist_Y [m]', 2)
```
